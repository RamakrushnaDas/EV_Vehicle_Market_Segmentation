```python
## WCSS with samples

from sklearn.utils import resample
from sklearn.cluster import KMeans


nboot = 100
wcss_1 = []
kmeans_models=[]

for k in range(2, 9):
    inertia_values = []
    for _ in range(nboot):
        bootstrap_sample = resample(df.values,random_state=1234)
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
        kmeans.fit(bootstrap_sample)
        inertia_values.append(kmeans.inertia_)


    wcss_1.append(np.mean(inertia_values))
    kmeans_models.append(kmeans)
```

```python
plt.plot(range(2,9), wcss_1,marker='o')
plt.xticks(range(2,9))
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
##With samples ARI
from sklearn.metrics import adjusted_rand_score
np.random.seed(1234)


nboot = 100
cluster_counts_1 = range(2, 9)
ari_scores_1 = []


for i in cluster_counts_1:
    kmeans_reff = KMeans(n_clusters=i, random_state=1234, n_init=10, init='k-means++')
    pseudo_labels = kmeans_reff.fit_predict(df.values)
    bootstrap_ari_scores = []

    for _ in range(nboot):
        bootstrap_samples = resample(df.values)
        kmeanss = KMeans(n_clusters=i, random_state=1234, n_init=10, init='k-means++')
        predicted_labels = kmeanss.fit_predict(bootstrap_samples)
        ari = adjusted_rand_score(pseudo_labels, predicted_labels)
        bootstrap_ari_scores.append(ari)

    ari_scores_1.append(bootstrap_ari_scores)


ari_scores_1 = np.array(ari_scores_1).T
```

```python

from sklearn.mixture import GaussianMixture
np.random.seed(1234)
num_clusters = range(2, 9)
n_rep = 10


best_models = []

for k in num_clusters:
    best_aic, best_bic, best_icl = np.inf, np.inf, np.inf

    for _ in range(n_rep):
        model = GaussianMixture(n_components=k, covariance_type='full', random_state=1234)
        model.fit(df)


        aic = model.aic(df)
        bic = model.bic(df)


        responsibilities = model.predict_proba(df)
        log_resp = np.log(responsibilities + 1e-10)
        icl = bic - np.sum(responsibilities * log_resp)


        if bic < best_bic:
            best_aic, best_bic, best_icl = aic, bic, icl

    best_models.append((k, best_aic, best_bic, best_icl))

for k, aic, bic, icl in best_models:
    print(f"Number of clusters: {k}, AIC: {aic}, BIC: {bic}, ICL: {icl}")
```

```python
import matplotlib.pyplot as plt


num_clusters = [model[0] for model in best_models]
aic_values = [model[1] for model in best_models]
bic_values = [model[2] for model in best_models]
icl_values = [model[3] for model in best_models]


plt.figure(figsize=(10, 6))
plt.plot(num_clusters, aic_values, marker='o', label='AIC', color='blue')
plt.plot(num_clusters, bic_values, marker='D', label='BIC', color='black')
plt.plot(num_clusters, icl_values, marker='*', label='ICL', color='red')


plt.xlabel("Number of Clusters (k)")
plt.ylabel("Value of Information Criteria (AIC, BIC, ICL)")
plt.title("Information Criteria vs. Number of Clusters")
plt.legend()
#plt.grid(True)
plt.show()
```
